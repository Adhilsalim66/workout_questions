1)Basic Array Operations

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

mpg = df["mpg"].to_numpy()

print("Mean MPG:", np.mean(mpg))
print("Median MPG:", np.median(mpg))
print("Standard Deviation:", np.std(mpg))

count = np.sum(mpg > 25)
print("Cars with MPG > 25:", count)
```

2)Filtering

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

cylinders = df["cylinders"].to_numpy()
car_names = df["car name"].to_numpy()

filtered_cars = car_names[cylinders > 6]

print("Cars with more than 6 cylinders:")
print(filtered_cars.tolist())
```

3)Broadcasting


```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

hp = pd.to_numeric(df["horsepower"], errors="coerce").to_numpy()

mean_hp = np.nanmean(hp)
hp[np.isnan(hp)] = mean_hp

updated_hp = hp * 1.10
print("Updated Horsepower Values:")
print(updated_hp)
```

4)Boolean Indexing

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

displacement = df["displacement"].to_numpy()
origin = df["origin"].to_numpy()

europe_disp = displacement[origin == 2]

avg_disp = np.mean(europe_disp)

print("Average displacement of European cars:", avg_disp)
```

5) Conditional Aggregates


```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

mpg = df["mpg"].to_numpy()
cylinders = df["cylinders"].to_numpy()

unique_cyl = np.unique(cylinders)

print("Mean MPG by Cylinders:")
for cyl in unique_cyl:
    mean_mpg = np.mean(mpg[cylinders == cyl])
    print(f"{cyl} cylinders:", mean_mpg)
```

 2)Filtering and Indexing


```python
import pandas as pd

df = pd.read_csv("datafile.csv")
df.columns = df.columns.str.replace(" ", "_")

filtered = df[(df["model_year"] == 75) & (df["weight"] < 3000)]

result = filtered[["car_name", "weight", "mpg"]]

print(result)
```

3)Handling Missing Data

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

print(df.isnull().sum())

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

median_hp = df["horsepower"].median()

df["horsepower"].fillna(median_hp, inplace=True)

print(df["horsepower"].isnull().sum())
```

4)Data Transformation


```python
import pandas as pd

df = pd.read_csv("datafile.csv")

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

df["power_to_weight_ratio"] = df["horsepower"] / df["weight"]

print(df[["horsepower", "weight", "power_to_weight_ratio"]].head())
```

5)Group By

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

result = df.groupby("origin")["mpg"].mean()

print(result)
```

6)Visualization Preparation

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

summary = df.groupby("model year")[["mpg", "weight", "horsepower"]].mean()

print(summary)
```

 7)Exporting Data

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

filtered = df[df["mpg"] > 30][["mpg", "cylinders", "horsepower", "weight"]]

filtered.to_csv("high_mpg_cars.csv", index=False)

print("File saved as high_mpg_cars.csv")
```
